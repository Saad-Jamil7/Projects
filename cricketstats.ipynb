Top 10 Test Batsmen

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def display_batsmen(df, format_name):

    filtered_df = df[df["Format"] == format_name].copy()
    
    filtered_df.index = range(1, len(filtered_df) + 1)
    
    print(f"Top 10 {format_name} Batsmen".center(80))

    print(filtered_df)
    print("\n")  

df = pd.read_csv("batsmen.csv")

display_batsmen(df, "Test")


```

Top 10 ODI Batsmen

```python
display_batsmen(df,"ODI")
```

Top T20I Batsmen


```python
display_batsmen(df,"T20I")
```

Stats For Test Innings & Average


```python

df = pd.read_csv("batsmen.csv")

test_batsmen=df[df["Format"]=="Test"]

melted = test_batsmen.melt(id_vars="Batsman", value_vars=["Innings","Average"], 
                       var_name="Statistic", value_name="Value")
plt.figure(figsize=(12,6))
sns.barplot(x="Batsman",y="Value",data=melted,hue="Statistic",palette="Set2")
plt.title("Top 10 Test Batsmen - Innings & Average", fontsize=16)
plt.xlabel("Batsman", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Statistic")
plt.show()
```

Consistency of Test Batsman(2020-24)

```python
import numpy as np
import pandas as pd

consistency = {}

for player in df.columns[1:]:
    runs = df[player].values
    median_runs = np.median(runs)
    
    q1 = np.percentile(runs, 25)
    q3 = np.percentile(runs, 75)
    iqr = q3 - q1
    
    consistency_score = iqr / median_runs if median_runs != 0 else 0
    
    consistency[player] = {
        "Median Runs": round(median_runs, 2),
        "IQR": round(iqr, 2),
        "Consistency Score": round(consistency_score, 2)
    }

consistency_df = pd.DataFrame(consistency).T
print(consistency_df)

consistency_sorted = consistency_df.sort_values("Consistency Score")

plt.figure(figsize=(10,6))
plt.plot(consistency_sorted.index, consistency_sorted["Consistency Score"], marker='o', linestyle='-', color='Green')
plt.xticks(rotation=45)
plt.ylabel("Consistency Score")
plt.title("Batsmen Consistency Score (Higher=Inconsistent)")
plt.grid(True)
plt.show()
```
